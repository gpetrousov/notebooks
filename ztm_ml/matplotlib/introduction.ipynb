# Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prep some data
x = [1, 2, 3, 4,]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. plot the data
ax.plot(x, y)

# 4. customize plot
ax.set(title="Simple plot",
      xlabel="x-axis",
      ylabel="y-label")

# 5. Save & show
fig.savefig("plots/sample-plot.png")
```

```python
# Histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Multiple plots at the same time - subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
# Not that the 2nd argument is a tupple so you can just use a single variable to create that tupple.

# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(x, x**2);
ax3.hist(x);
ax4.bar(prices.keys(), prices.values())
```

```python
# Using the Object  Oriented method is the preferred way if we want to make more complicated plots.
```

```python
heart_disease = pd.read_csv("data/heart-disease.csv")
heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
## OO method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind = "scatter",
            x = "age",
            y = "chol",
            c = "target",
            ax = ax);
# ax.set_xlim([45, 100])
```

```python
## OO method from sctratch
fig, ax = plt.subplots(figsize=(10,6))

scatter  = ax.scatter(
    x=over_50["age"],
    y=over_50["chol"],
    c=over_50["target"]
)

# CUstomize the above plot
ax.set(title="Heart disease & Cholesterol levels",
      xlabel="Age",
      ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line

ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10));

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],y=over_50["chol"], c=over_50["target"])

# Customize ax0
ax0.set(title="Heart disease & cholesterol levels", xlabel="Age", ylabel="Cholesterol");

# Add legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a meanline
ax0.axhline(over_50["chol"].mean(), linestyle="--")

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],y=over_50["thalach"], c=over_50["target"])
ax1.set(title="Heart disease & Max heart rate (thalach)", xlabel="Age", ylabel="Cholesterol");
ax1.legend(*scatter.legend_elements(), title="Target");
ax1.axhline(over_50["thalach"].mean(), linestyle="--");

fig.text(x=0,y=0,s="Heart disease analysis");
```

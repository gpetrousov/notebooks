## Dot product example (nut butter sales)

```python
np.random.seed(0)
# Nof jars sold
sales_amounts = np.random.randint(20, size=(5,3))
sales_amounts
```

```python
# Creaete weekly_sales DataFrame
weekly_sales = pd.DataFrame(sales_amounts,
                            index=["Mon", "Tue", "Wed", "Thu", "Fri"],
                            columns=["Almond butter", "Peanut butter", "Cashew butter"])
weekly_sales
```

```python
# Create the prices array
prices = np.array([10, 8, 12])
prices, prices.shape
```

```python
# Create butter prices DataFrame
# Watch how we need to reshape the prices array because of how pandas handles rows and columns differently.
butter_prices = pd.DataFrame(prices.reshape(1,3),
                            index=["Price"],
                            columns=["Almond butter", "Peanut butter", "Cashew butter"])
butter_prices
```

```python
total_sales = prices.dot(sales_amounts)
```

```python
# Shapes are not aligned, we need to transpose
total_sales = prices.dot(sales_amounts.T)
total_sales
```

```python
# Create daily saleds
butter_prices.shape, weekly_sales.shape
```

```python
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
weekly_sales
```

```python
# Add the totals into the weekly sales
weekly_sales["Total ($)"] = daily_sales.T
weekly_sales
```
